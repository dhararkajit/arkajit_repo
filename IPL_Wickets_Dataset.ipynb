```python
df_bowlers = df_ms_wickets[df_ms_wickets['Inns'].astype(float) > 0.0]
```

```python
cols = ['Wkts','Inns','Overs','Runs','Mdns','Ave','Econ','SR']
df_bowlers[cols] = df_bowlers[cols].astype(float)
```

```python
df_bowlers.reset_index(drop=True , inplace = True)
```

```python
df_bowlers.drop(df_bowlers[df_bowlers['team'] == 'DC'].index ,inplace= True)
df_bowlers.drop(df_bowlers[df_bowlers['team'] == 'PW'].index , inplace= True)
```

```python
def get_bowling_statistics(df,parameter, statistic,output):
    df = df.groupby(['year','team','status'], as_index='False').agg({parameter: statistic}).apply(lambda x : round(x,2))
    df.reset_index(level = ['team','year','status'], inplace=True)
    return df.pivot(columns='team', values= output, index='year')
```

# Qualification Status of all the Teams

```python
df =get_bowling_statistics(df_bowlers,'Wkts','max','status')
print (df)
```

# Mean/Max/Min Wickets taken of the bowlers of all Teams

```python
df=get_bowling_statistics(df_bowlers[df_bowlers['Wkts'] > 0.0],'Wkts','mean','Wkts')
print (df)
axes = df.plot(kind = 'line',grid = True,subplots = True , layout=(5,3), figsize = (15,20) )
```

# Mean/Max/Min Bowling Average of the bowlers of all Teams

```python
df=get_bowling_statistics(df_bowlers[df_bowlers['Wkts'] > 0.0],'Ave','mean','Ave')
print (df)
axes = df.plot(kind = 'line',grid = True,subplots = True , layout=(5,3), figsize = (15,20) )
```

# Mean/Max/Min Bowling Strike Rate of the bowlers of all Teams

```python
df=get_bowling_statistics(df_bowlers[df_bowlers['Wkts'] > 0.0],'SR','mean','SR')
print (df)
axes = df.plot(kind = 'line',grid = True,subplots = True , layout=(5,3), figsize = (15,20) )
```

# Number of different bowlers used by all Teams

```python
df=get_bowling_statistics(df_bowlers,'Inns','count','Inns')
print (df)
axes = df.plot(kind = 'line',grid = True,subplots = True , layout=(5,3), figsize = (15,20) )
```
