```python
import pandas as pd 
import numpy as np
import glob
#import matplotlib.pyplot as plt
```

```python
path =r'C:\Continuum\Anaconda2\envs\TFENV\IPL_TEAMS_STAT' # use your path
allFiles = glob.glob(path + "/*4n6*.csv")
df_4n6_stats = pd.DataFrame()
list_ = []
for file_ in allFiles:
    df = pd.read_csv(file_,index_col=0)
    year = (file_.rsplit('.csv', 1)[0]).rsplit('_', 1)[1]
    df['year'] = year
    list_.append(df)
df_4n6_stats = pd.concat(list_)
```
