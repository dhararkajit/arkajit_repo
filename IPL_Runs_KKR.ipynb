```python
from bs4 import BeautifulSoup
import urllib
import pandas as pd
import numpy as np
```

```python
def make_list(rows):
    list_data = []
    head =[row.find_all('th') for row in rows]
    list_data.append([ele.text.strip() for ele in head[0]])
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        list_data.append([ele for ele in cols if ele !=[]])
    return list_data
```

```python
m_url = r'http://www.iplt20.com/stats/2017'
m_2017 = urllib.request.urlopen(m_url)
m_html_2017 = BeautifulSoup(m_2017)
m_rows = m_html_2017.find_all('tr')
```

```python
list_m_2017 = make_list(m_rows)
```

```python
df_m_2017 = pd.DataFrame(list_m_2017[2:], columns = list_m_2017[0])
```

```python
url_most_runs = r'http://stats.espncricinfo.com/indian-premier-league-2017/engine/records/batting/most_runs_career.html?id=11701;type=tournament'
most_runs_2017 = urllib.request.urlopen(url_most_runs)
mr_html_2017 = BeautifulSoup(most_runs_2017)
mr_rows = mr_html_2017.find_all('tr')
```

```python
list_mr_2017 = make_list(mr_rows)
```

```python
df_mr_2017 = pd.DataFrame(list_mr_2017[2:102], columns = list_mr_2017[0])
```

```python
df_mr_2017.to_csv('Most_runs_2017.csv')
```
