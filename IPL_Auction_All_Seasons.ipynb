```python
from bs4 import BeautifulSoup
import urllib
import pandas as pd
import numpy as np
```

```python
def make_list(rows):
    list_data = []
    head =[row.find_all('th') for row in rows]
    list_data.append([ele.text.strip() for ele in head[0]])
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        list_data.append([ele for ele in cols if ele !=[]])
    return list_data
```

```python
url_auction_2008='https://en.wikipedia.org/wiki/2008_Indian_Premier_League'
auction_2008 = urllib.request.urlopen(url_auction_2008)
auc_html_2008 = BeautifulSoup(auction_2008,'lxml')
auc_rows = auc_html_2008.find('table',attrs={'class':'wikitable sortable'})
auc_rows = auc_rows.find_all('tr')
```

```python
list_auc_2008 = make_list(auc_rows)
```

```python
df_auc_2008 = pd.DataFrame(list_auc_2008[2:], columns = list_auc_2008[0])
```

```python
df_auc_2008
```
